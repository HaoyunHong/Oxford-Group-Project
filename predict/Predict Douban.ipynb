```python
movie_genres = dict(zip(douban_all['电影'], douban_all['类型']))
```

```python
from collections import OrderedDict

douban_review = douban_all[['电影', '类型']]

movie_review = OrderedDict()

for index, row in douban_df.iterrows():
    name = row['电影']
    if not movie_review.__contains__(name):
        movie_review[name] = ''
    movie_review[name] += ' '+str(row['评论'])

douban_review['review_sum'] = list(movie_review.values())
douban_review
```

```python
# random split train and test sets 8:2
import random

random.seed(37)

all_indices = range(0,250)
train_indices = random.sample(all_indices, 200)
train_names = douban_review['电影'].iloc[train_indices]

train_names
```

```python
test_indices = [i for i in all_indices if i not in train_indices]

test_names = douban_review['电影'].iloc[test_indices]

test_names
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

tfidf = TfidfVectorizer(stop_words='english')

matrix = tfidf.fit_transform(douban_review['review_sum'])
```

```python
cosine_sim = linear_kernel(matrix, matrix)
```

```python
indices = pd.Series(douban_review.index, index=douban_review['电影']).drop_duplicates()
```

```python
def get_top_similar(title, cosine_sim=cosine_sim):
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    true_gen = douban_review['类型'].iloc[sim_scores[0][0]]
    print("ground truth: ", true_gen)
    
    cand = []
    for i in sim_scores:
        if douban_review['电影'].iloc[i[0]] in list(train_names):
            cand.append(i[0])
    
    # top 10 similar in train data
    cand = cand[:10]
    
    
    genres_list = douban_review['类型'].iloc[cand]
    genres = []
    for g in genres_list:
        l = g.strip().split('/')
        for i in l:
            genres.append(i)
    genres = Counter(genres)
    
    # top 1 genres
    ans = genres.most_common(1)
    
    for a in ans:
        prob = int(a[1])/10
        ans_gen = a[0]

    print("predicte: ", ans_gen, " (with confidence of ", prob, ")")
    
    true_gen = true_gen.strip().split('/')
    if ans_gen in true_gen:
        return 1
    else:
        return 0
```

```python
hit1 = 0

for name in test_names:
    print("Test movie: ", name)
    res = get_top_similar(name)
    hit1 += res
    print("********************************************************************")
```

```python
hit1 = hit1/len(test_names)
print("predict accuracy is: ", hit1)
```
